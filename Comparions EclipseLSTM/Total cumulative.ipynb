```python
import pandas as pd
```

```python
df = pd.read_excel("Book1.xlsx")
```

```python
df.head()
```

```python
import plotly.graph_objs as go
import plotly.offline as py
py.init_notebook_mode(connected = True)
```

```python
cum1 = []
x = 0
for i in range(len(df)):
    x += df['CUM'].iloc[i]
    cum1.append(x)
```

```python
trace = go.Scatter(x = df.DATEPRD,
                  y = cum1,
                  name = "Actual value")
data = [trace]
layout = go.Layout(title = "Comparision of oil production on well F1")
fig = go.Figure(data = data, layout = layout)
py.iplot(fig)
```

```python
df2 = pd.read_excel("Book2.xlsx")
```

```python
df2.head()
```

```python
df3 = pd.read_excel("Totaleclipse.xlsx")
```

```python
cum3 = []
x = 0
for i in range(len(df3)):
    x += df3['Total'].iloc[i]
    cum3.append(x)
```

```python
cum2 = []
x = 0
for i in range(len(df2)):
    x += df2['Total'].iloc[i]
    cum2.append(x)
    
```

```python
trace = go.Scatter(x = df.DATEPRD,
                  y = cum1,
                  name = "Total LSTM")
trace1 = go.Scatter(x = df2.F12,
                   y = cum2,
                   name = "Total actual values")
trace3 = go.Scatter(x = df3.Date2,
                   y = cum3,
                   name = "Total Eclipse")
data = [trace1, trace, trace3]
layout = go.Layout(title = "Comparision of cumulative oil production on well F12")
fig = go.Figure(data = data, layout = layout)
py.iplot(fig)
```

```python
trace = go.Scatter(x = df.DATEPRD,
                  y = df.CUM,
                  name = "Total predicted by LSTM")
trace1 = go.Scatter(x = df2.F12,
                   y = df2.Total,
                   name = "Total oil production")
trace3 = go.Scatter(x = df3.Date2,
                   y = df3.Total/10,
                   name = "Total predicted by Eclipse")
data = [trace1, trace, trace3]
layout = go.Layout(title = "Comparision of oil production on well F1")
fig = go.Figure(data = data, layout = layout)
py.iplot(fig)
```
