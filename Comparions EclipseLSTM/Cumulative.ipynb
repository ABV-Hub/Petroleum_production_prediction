```python
def cum(data):
    cum1 = []
    cum2 = []
    x ,y = 0,0
    for i in range(len(data)):
        x += data['Production'][i]
        y += data['OIL_PRE'][i]
        cum1.append(x)
        cum2.append(y)
    return cum1, cum2
    
```

```python
cum1, cum2 = cum(df3)
```

```python
trace = go.Scatter(x = df3.Date1,
                  y = cum1,
                  name = "Actual values")
trace1 = go.Scatter(x = df3.DATEPRD,
                   y = cum2,
                   name = "LSTM")
data = [trace, trace1]
layout = go.Layout(title = "Comparision of cumulative oil production on well F12")
fig = go.Figure(data = data, layout = layout)
py.iplot(fig)
```
